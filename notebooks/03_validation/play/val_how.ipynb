```python
def check(code, expected):
    try:
        output = eval(code) 
        valid = output == expected
    except BaseException as exc:
        output = exc
        valid = isinstance(exc, expected) if isinstance(expected, Exception) else False
    
    valid_str = "✅" if valid else "❌"
    output_str = output if not isinstance(output, Exception) else type(output).__name__
    expected_str = expected.__name__ if isinstance(expected, type) and issubclass(expected, Exception) else expected
    print(valid_str, code, "->", output_str, "" if valid else f"(Expected: {expected_str})")
        
```

```python
def add(x, y=None):
    return x + y
```

```python
check("add(2, 3)", 5)
check("add(3)", 3)
check("add('1', 3)", TypeError)
check("add(2, True)", TypeError)
check("add('1', '4')", TypeError)
check("add([1, 2, 3])", 6)
```

```python
nts = {'g': 'c', 'c': 'g', 'a': 'u', 't': 'a'}

def transcribe(dna):
    dna_r = dna[::-1]
    rna = ''.join(nts[nt] for nt in dna_r)
    return rna
    
```

```python
check("transcribe('agtc')", 'gacu')
check("transcribe('munich')", ValueError)
check("transcribe(['g', 't', 'c'])", TypeError)
# check("")
```
